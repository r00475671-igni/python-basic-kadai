# if文を用いて、以下の条件分岐をするコードを記述してください。
# 値はprint関数で出力してください。


*   変数varが、3の倍数の場合は「Fizz」を出力
*   変数varが、5の倍数の場合は「Buzz」を出力
*   変数varが、3の倍数と5の倍数の両方に該当する場合は「FizzBuzz」を出力
*   上記のどの場合にも該当しない場合は、変数varの値を出力


ただし、変数varは正の整数とします。

```python
var= 28
```

```python
if var % 3 == 0 and var % 5 == 0:
  print("FizzBuzz")
elif var % 5 == 0:
  print("Buzz")
elif var % 3 == 0:
  print("Fizz")
else:
  print(var)
```
