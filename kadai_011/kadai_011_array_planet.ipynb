```python
#リストの中身を、先頭からすべて取り出して出力する
```

```python
array = ["水","金","地","火","木","土","天","海","冥"]
```

```python
#for文で出力
for i in range(len(array)):
    print(array[i])
```

```python
#for文で出力
for planet_name in array:
    print(planet_name)
```

```python
#while文で出力

i = 0

while i < len(array):
    print(array[i])
    i += 1


```
